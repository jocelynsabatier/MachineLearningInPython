```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
link = "https://raw.githubusercontent.com/murpi/wilddata/master/pokemon.csv"
pokemon = pd.read_csv(link)
```

```python
pokemon
```

```python
pokemon = pokemon[['Name', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Legendary']]
pokemon
```

```python
non_legendary_pokemon = pokemon[~(pokemon['Legendary'] == True)]
```

```python
non_legendary_pokemon
```

```python
legendary_pokemon = pokemon[(pokemon['Legendary'] == True)]
```

```python
legendary_pokemon
```

```python
#on cherche a trouvÃ© les voisin de Mewtwo, Lugia, Rayquaza, Giratina, Dialga, et Palkia.

#commenÃ§ons par trouver les stat de Mewtwo :

mewtwo = legendary_pokemon[(legendary_pokemon['Name'] == 'Mewtwo')]
mewtwo
```

```python

from sklearn.neighbors import NearestNeighbors

X = non_legendary_pokemon[['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

distanceKNN = NearestNeighbors(n_neighbors=1).fit(X)

distanceKNN.kneighbors([[106,	110,	90,	154,	90,	130]])
```

```python
neighbors = distanceKNN.kneighbors([[106,	110,	90,	154,	90,	130]])

non_legendary_pokemon.iloc[neighbors[1][0][0]]


#le meilleur remplaÃ§ant pour mewtwo est Mega Houndoom
```
